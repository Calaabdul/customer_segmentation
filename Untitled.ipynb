### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

### Retrieving and exploring data

```python
data = pd.read_csv('data/segmentation data.csv')
data.head()
```

```python
data.info()
```

Data has **no Missing nor Duplicate values**

```python
train, test = train_test_split(data, random_state = 2025, test_size = .20)
```

```python
print(train.shape)
print(test.shape)
```

### Find best k(Number of Clusters)

Using the Elbow Method, Find the best number of clusters

```python
ks = range(1, 8)
inertia = []
score = []
for k in ks:
    cl_model = Pipeline(steps = [
        ('scaler',StandardScaler()),
        ('cluster', KMeans(n_clusters = k, random_state = 2025))])
    cl_model.fit(train)
    inertia.append(cl_model['cluster'].inertia_)
    
print(inertia)
```

```python
plt.plot(ks, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show();
```

It appears **3** is the ideal number of clusters for Customers

## Evaluate Model

```python
k_model = KMeans(n_clusters = 3, random_state = 2025)
labels = k_model.fit_predict(train)
labels
```

```python
new_labels = k_model.predict(test)
```

```python
plt.scatter(test['Age'], test['Income'], c = new_labels, alpha = .7);
```
